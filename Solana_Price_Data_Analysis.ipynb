## Feature engineering

### Subtask:
Engineer new features based on the insights gained from the data analysis and visualization.


**Reasoning**:
Create new features based on the insights gained from the data analysis and visualization, including interaction terms, polynomial features, rolling window features, binary features, and domain-specific features.



```python
df_engineered = df_copy.copy()

# 1. Interaction terms: RSI and MACD interaction
df_engineered['RSI_MACD_Interaction'] = df_engineered['RSI_14'] * df_engineered['MACD_12_26_9']

# 2. Polynomial features: Quadratic term for RSI
df_engineered['RSI_Squared'] = df_engineered['RSI_14'] ** 2

# 3. Rolling window features: 7-day rolling mean of closing price
df_engineered['Close_RollingMean_7'] = df_engineered['Close'].rolling(window=7).mean()

# 4. Binary features: Closing price above 50-day moving average
df_engineered['Close_Above_SMA_50'] = (df_engineered['Close'] > df_engineered['SMA_50']).astype(int)

# 5. Domain-specific feature: Price change percentage
df_engineered['PriceChangePercentage'] = (df_engineered['Close'] - df_engineered['Close'].shift(1)) / df_engineered['Close'].shift(1) * 100


import numpy as np

# Replace infinite values with NaN explicitly
df_engineered['PriceChangePercentage'] = df_engineered['PriceChangePercentage'].replace([np.inf, -np.inf], np.nan)

# Apply backward fill inplace
df_engineered.bfill(inplace=True)

# Display the first few rows
display(df_engineered.head())

df_engineered.to_csv('df_testing_engineered1.csv', index=False)
```

Handle the `inf` values in the `PriceChangePercentage` column, which likely resulted from a division by zero.

**Reasoning**:
The previous code block still had chained assignment warnings. We need to rewrite the code to avoid these warnings and use `.bfill()` to fill NaN values instead of `fillna(method='bfill')`.



```python
import numpy as np

df_engineered['PriceChangePercentage'] = df_engineered['PriceChangePercentage'].replace([np.inf, -np.inf], np.nan)
df_engineered['PriceChangePercentage'] = df_engineered['PriceChangePercentage'].bfill()

display(df_engineered.head())
```

## Data splitting

### Subtask:
Split the engineered dataset `df_engineered` into training, validation, and testing sets, ensuring that the time series nature of the data is preserved.


**Reasoning**:
Split the data into training, validation, and testing sets while preserving the time series order.



```python
from sklearn.model_selection import train_test_split

# Define the target variable
y = df_engineered['Close']
X = df_engineered.drop(['Close', 'date'], axis=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, shuffle=False
)
```

## Model training

### Subtask:
Train a Linear Regression model to predict the closing price of Solana.


**Reasoning**:
Train a linear regression model using the training data and make predictions on the validation set.



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Instantiate a LinearRegression object
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred_val = model.predict(X_val)

# Print the coefficients and R-squared score
print("Coefficients:", model.coef_)
print("R-squared (training):", r2_score(y_train, model.predict(X_train)))
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained Linear Regression model on the validation set.


**Reasoning**:
Evaluate the trained Linear Regression model on the validation set by calculating the MAE, MSE, RMSE, and R-squared scores.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Predict the closing price on the validation set
y_pred_val = model.predict(X_val)

# Calculate the evaluation metrics
mae = mean_absolute_error(y_val, y_pred_val)
mse = mean_squared_error(y_val, y_pred_val)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred_val)

# Print the results
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```

## Model optimization

### Subtask:
Optimize the Linear Regression model by performing hyperparameter tuning using cross-validation.


**Reasoning**:
Import necessary libraries and perform hyperparameter tuning using GridSearchCV with 5-fold cross-validation for the Linear Regression model.



```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {'fit_intercept': [True, False]}

# Create a Linear Regression model
model = LinearRegression()

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and the best score
print("Best parameters:", grid_search.best_params_)
print("Best score (negative MSE):", grid_search.best_score_)

# Get the best model
best_model = grid_search.best_estimator_

# Evaluate the best model on the validation set
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred_val = best_model.predict(X_val)
mae = mean_absolute_error(y_val, y_pred_val)
mse = mean_squared_error(y_val, y_pred_val)
rmse = np.sqrt(mse)
r2 = r2_score(y_val, y_pred_val)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```
