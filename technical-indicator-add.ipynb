```python
import pandas as pd
import pandas_ta as ta
```

```python
df = pd.read_csv('solana_testing_data.csv')

# Add RSI, MACD, etc.
df.ta.rsi(length=14, append=True)       # Adds 'RSI_14' column
df.ta.macd(append=True)                 # Adds 'MACD_12_26_9', 'MACDh_12_26_9', etc.
df.ta.sma(length=50, append=True)       # Adds 'SMA_50'
df.ta.bbands(append=True)               # Adds 'BBL_5_2.0', 'BBM_5_2.0', etc.

df.to_csv('solana_testing_with_indicators.csv', index=False)
```
